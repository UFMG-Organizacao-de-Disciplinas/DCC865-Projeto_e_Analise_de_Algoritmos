A tarefa é simples. Através de alguns pontos críticos em 2D, você deve desenhar uma onda como uma curva. Seu objetivo é incluir tantos pontos quantos forem possível.

- Haverá uma linha imaginária $y = a$, a qual chamaremos de eixo principal para a curva.
- Todos os pontos na curva deverão ter diferentes coordenadas $x$. Suas coordenadas $y$ devem ser na forma $a-1$ ou $a+1$.

Dois pontos consecutivos na curva devem ter diferença de $2$ na coordenada $y$.

## Entrada (Onda Crítica)

Haverá no máximo, $222$ casos de testes. Cada caso inicia com um inteiro $N$, que é o número de pontos no caso de teste. Nas próximas $N$ linhas, haverá $N$ pares de inteiros dando as coordenadas $x$ e $y$ de cada ponto. Não haverá mais do que $1000$ pontos em cada caso de teste. Todas coordenadas são inteiros - eles devem ficar dentro de um inteiro de $2$ bytes com sinal. Os dados devem ser lidos da entrada padrão.

## Saída (Onda Crítica)

Para cada caso de teste, imprima um número - o número máximo de pontos críticos que podem ser incluídos em uma curva desenhada a partir dos pontos dados.

```python
"""  """

def get_max_traversal(origin, destiny):
    quantidade_de_pontos = 1
    i = 0
    j = 0

    current_direction = -1
    while i < len(origin) and j < len(destiny):
        if current_direction == -1:
            while j < len(destiny) and destiny[j] < origin[i]:
                j += 1
            if j < len(destiny) and destiny[j] > origin[i]:
                quantidade_de_pontos += 1
                current_direction = 1
        elif current_direction == 1:
            while i < len(origin) and origin[i] < destiny[j]:
                i += 1
            if i < len(origin) and origin[i] > destiny[j]:
                quantidade_de_pontos += 1
                current_direction = -1

    return quantidade_de_pontos


def get_max_traversal_2(origin, destiny):
    quantidade_de_pontos = 1
    i = 0
    j = 0
    current_direction = -1

    while i < len(origin) and j < len(destiny):
        if current_direction == -1:
            if destiny[j] > origin[i]:
                quantidade_de_pontos += 1
                current_direction = 1
            else:
                j += 1
        elif current_direction == 1:
            if origin[i] > destiny[j]:
                quantidade_de_pontos += 1
                current_direction = -1
            else:
                i += 1

    return quantidade_de_pontos


def get_pontos(N):
    pontos = dict()
    for i in range(N):
        x, y = map(int, input().split()) # X, Y = 2 byte signed int
        # x, y = map(int, dados.pop(0).split())  # X, Y = 2 byte signed int
        if pontos.get(y) is None:
            pontos[y] = [x]
        else:
            pontos[y].append(x)
    
    for key in pontos.keys():
        pontos[key] = sorted(pontos[key])
    
    return pontos
    
def get_instance_solution(pontos):
    global_max = 1
    sorted_keys = sorted(list(pontos.keys()))
    # print(pontos)
    # print(sorted_keys)
    for key in sorted_keys:
        if pontos.get(key + 2) is None:
            continue

        origin = pontos[key]
        destiny = pontos[key + 2]

        OD = get_max_traversal_2(origin, destiny)
        DO = get_max_traversal_2(destiny, origin)
        
        # print(f'({key} <--> {key + 2}): {OD} | {DO}') # debug
        global_max = max(global_max, DO, OD)

    return global_max

N = int(input()) # N <= 222; Número de casos de teste
# N = int(dados.pop(0)) if len(dados) > 0 else 0

while N != 0: # Segundo a Sabela: "confia"
    try:
        # pontos = [] # len(pontos) <= 1000; Lista de pontos
        pontos = get_pontos(N)
        instance_solution = get_instance_solution(pontos)
        print(instance_solution)
        
        N = int(input())
        # N = int(dados.pop(0))
    except EOFError:
        break
    except IndexError:
        break
```
