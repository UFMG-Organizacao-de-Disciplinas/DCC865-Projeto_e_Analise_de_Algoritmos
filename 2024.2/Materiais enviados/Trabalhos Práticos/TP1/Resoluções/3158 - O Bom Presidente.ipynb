Livrolândia é um país que, como o nome já diz, preza pela leitura. Nesta cidade há uma regra universal: toda cidade do país deve ter acesso a bibliotecas. Todos os presidentes que passaram por Livrolândia conseguiram manter esta regra.

Roci é o atual presidente, e fez questão de dar manutenção a todas as bibliotecas do país, além de manter a boa qualidade das estradas entre as cidades, para que cidades que não tem biblioteca, consigam acesso a cidades vizinhas que tenham.

Infelizmente, Roci é muito azarado e, logo em seu mandato, um tornado destruiu todas as bibliotecas e obstruiu todas as estradas de Livrolândia. Agora, o presidente terá que bolar um plano para reconstruir o país, seguindo sua regra universal e com o menor custo possível para as obras.

Livrolândia tem $N$ cidades numeradas de $1$ a $N$. As cidades são conectadas por $M$ estradas bidirecionadas. Uma cidade tem acesso a uma biblioteca se:

- Esta cidade tem uma biblioteca;
- É possível, a partir desta cidade, viajar para uma cidade que contém uma biblioteca.

O custo para reparar uma estrada é $E$ tolkiens (tolkiens é a moeda de Livrolândia) e o custo para construir uma biblioteca é $B$ tolkiens.

Dado o mapa de Livrolândia e os custos de reparo e construção, escreva um programa que retorne o custo mínimo para reconstruir o país, seguindo a regra universal, e assim, salve Roci.

## Entrada - beecrowd | 3158

A primeira linha da entrada contém um inteiro $T$ indicando o número de possíveis mapas.

A segunda linha da entrada contém 4 inteiros, $N$, $M$, $B$ e $E$, número de cidades, número de estradas, custo para construir uma biblioteca e o custo para construir uma estrada, respectivamente.

Depois há $M$ linhas indicando as estradas obstruídas, em que cada uma há dois inteiros $X$ e $Y$, indicando que há uma estrada que liga a cidade $X$ à cidade $Y$.

Limites:

- $1 \leq T \leq 10;$
- $1 \leq N \leq 10^5;$
- $0 \leq M \leq \min(10^5, (N*(N-1))/2);$
- $1 \leq  B, E \leq 10^5;$
- $1 \leq X,Y \leq N.$

## Saída - beecrowd | 3158

Para cada possível mapa, indique o custo mínimo para reconstruir o país.

```python
""" Input """


def get_input(bc_run=False):
    """ Get input from user """
    
    def get_beecrowd_input(input_data, index, is_running_on_beecrowd=False):
        return input() if is_running_on_beecrowd else input_data[index]
    
    result_check = [4, 12]
    input_data = [
        "2",
        "3 3 2 1",
        "1 2",
        "3 1",
        "2 3",
        "6 6 2 5",
        "1 3",
        "3 4",
        "2 4",
        "1 2",
        "2 3",
        "5 6",
    ]

    def convert_input_to_list_of_integers(input_data_item):
        """ Convert input to list of integers """
        list_of_integers = list(map(int, input_data_item.split()))
        # list_of_integers = [int(x) for x in input_data.split()]
        return list_of_integers

    index = 0

    T = convert_input_to_list_of_integers(get_beecrowd_input(input_data, index, bc_run))
    possible_maps = T[index]

    maps = []
    for _ in range(possible_maps):
        index += 1
        number_list = convert_input_to_list_of_integers(get_beecrowd_input(input_data, index, bc_run))
        if len(number_list) == 4:
            [N, M, B, E] = number_list
            map_info = {
                "number_of_cities": N,
                "number_of_roads": M,
                "cost_of_library": B,
                "cost_of_road": E,
                "roads": []
            }
            for _ in range(map_info["number_of_roads"]):
                index += 1
                [X, Y] = convert_input_to_list_of_integers(get_beecrowd_input(input_data, index, bc_run))
                map_info["roads"].append([X, Y])
        maps.append(map_info)

    data_input = {
        "maps": maps,
        "result_check": result_check,
    }

    return data_input

def bom_presidente(input_dados):
    """ O Bom Presidente """
    def convert_list_of_edges_to_adjacency_list(edges, number_of_vertices):
        """ Convert list of edges to adjacency list """
        adjacency_list = [[] for _ in range(number_of_vertices)]
        for edge in edges:
            adjacency_list[edge[0] - 1].append(edge[1] - 1)
            adjacency_list[edge[1] - 1].append(edge[0] - 1)
        return adjacency_list
    
    def DFS(adjacency_list, number_of_vertices):
        """ Depth First Search """
        visited = [False] * number_of_vertices
        number_of_built_libraries = 0
        
        def DFS_visit(vertex):
            """ Depth First Search Visit """
            visited[vertex] = True
            for neighbor in adjacency_list[vertex]:
                if not visited[neighbor]:
                    DFS_visit(neighbor)
        
        for vertex in range(number_of_vertices):
            if not visited[vertex]:
                DFS_visit(vertex)
                number_of_built_libraries += 1
        
        return number_of_built_libraries
    
    def get_graph_values(map):
        """ Get graph values """
        number_of_built_libraries = 0
        adjacency_list = convert_list_of_edges_to_adjacency_list(map["roads"], map["number_of_cities"])
        
        number_of_built_libraries = DFS(adjacency_list, map["number_of_cities"])
        
        return number_of_built_libraries
    
    def get_city_rebuild_minimun_cost(map):
        """ Get city rebuild minimun cost """
        minimun_cost = 0
        if map["cost_of_library"] <= map["cost_of_road"]:
            minimun_cost = map["cost_of_library"] * map["number_of_cities"]
        else:
            number_of_built_libraries = get_graph_values(map)
            number_of_built_roads = map["number_of_cities"] - number_of_built_libraries
            minimun_cost = number_of_built_libraries * map["cost_of_library"] + number_of_built_roads * map["cost_of_road"]
        return minimun_cost
    
    result = []
    for map in input_dados["maps"]:
        result.append(get_city_rebuild_minimun_cost(map))
    return result

def format_output(output_dados):
    """ Format output """
    return "\n".join(map(str, output_dados))

def main():
    """ Main function """
    input_dados = get_input(bc_run=False)
    # print(input_dados)
    output_dados = bom_presidente(input_dados)
    formatted_output = format_output(output_dados)
    print(formatted_output)

main()
```
