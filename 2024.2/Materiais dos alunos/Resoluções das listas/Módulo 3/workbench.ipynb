```python
def get_best_pair(P):
    n = len(P)
    i = 0
    j = 0
    min_i = 0
    max_diff = 0
    for k in range(1, n):
        if P[k] < P[min_i]:
            min_i = k
        diff = P[k] - P[min_i]
        if diff > max_diff:
            max_diff = diff
            i = min_i
            j = k
    return f'If there was no way to make money during the {n} days' if i >= j else f'buy on {i+1}, sell on {j+1}'


print(get_best_pair([1, 2, 3, 4, 5]))
print(get_best_pair([3, 5, 1, 3, 4]))
print(get_best_pair([5, 4, 3, 2, 1]))
```

```python
""" kleinberg_mean """

def kleinberg_mean(n, a, b):
    k = n//2 + 1
    if n == 1:
        return min(A(a+k), B(b+k))
    if A(a+k) < B(b+k):
        return kleinberg_mean(k, a+k - 1, b)
    return kleinberg_mean(k, a, b+k - 1)

# kleinberg_mean(len(test_case), 0, 0)
```
