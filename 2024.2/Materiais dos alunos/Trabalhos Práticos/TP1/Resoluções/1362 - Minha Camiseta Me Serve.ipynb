# [beecrowd | 1362 - Minha Camiseta Me Serve](https://judge.beecrowd.com/pt/problems/view/1362)

Por Ginés García Mateos, UM  Espanha

Timelimit: 1

Nosso amigo Victor é instrutor em um programa ambiental voluntário. O chefe de Victor pediu para ele distribuir $N$ camisetas para $M$ voluntários ($N$ é múltiplo de seis, e $N \geq M$). Cada voluntário deve receber exatamente uma camiseta (se $N \neq M$, algumas camisetas podem sobrar). Há o mesmo número de camisetas disponíveis para cada tamanho de camiseta possível: XXL, XL, L, M, S e XS (siglas em inglês para P, M, G, etc.). Victor tem um pequeno problema: apenas dois tamanhos de camisetas servem para cada voluntário.

Você deve escrever um programa que decide se Victor pode distribuir as camisetas de tal forma que todo voluntário tenha uma camiseta que lhe serve.

## Entrada - beecrowd | 1362

A primeira linha da entrada contém o número de casos de teste.

Para cada caso de teste, há uma linha contendo os números $N$ e $M$. O número $N$ é múltiplo de seis, $1 \leq N \leq 36$, e indica o número total de camisetas disponíveis. O número $M$, $1 \leq M \leq 30$, indica o número de voluntários, com $N \geq M$. As próximas $M$ linhas descrevem os voluntários, um por linha. Cada linha contém dois tamanhos de camiseta possíveis (XXL, XL, L, M, S ou XS) separados por um espaço, indicando quais tamanhos servem para o voluntário.

## Saída - beecrowd | 1362

Para cada caso teste, imprima uma linha contendo YES se existe pelo menos uma maneira de distribuir as camisetas de tal forma que todo voluntário tenha uma camiseta que lhe serve, ou NO caso contrário.

```python
""" Minha Camiseta Me Serve """

def get_input(bc_run=False, default_test_case='dardoazul2'):
    """ Get input from user """
    def get_beecrowd_input(input_data, index, is_running_on_beecrowd=False):
        return input() if is_running_on_beecrowd else input_data[index]

    global test_cases
    current_test_cases = {'input': [], 'output': []
                          } if bc_run else test_cases[default_test_case]
    # current_test_cases = {'input': [], 'output': []} # if bc_run else test_cases['dardoazul2']

    index = 0
    input_test_cases = current_test_cases['input']
    number_of_test_cases = int(get_beecrowd_input(input_test_cases, index, bc_run))

    shirt_size_map = {'XS': 0, 'S': 1, 'M': 2, 'L': 3, 'XL': 4, 'XXL': 5}

    structured_test_cases = []

    for _ in range(number_of_test_cases):
        index += 1
        N, M = get_beecrowd_input(input_test_cases, index, bc_run).split()
        num_shirts = int(N)
        num_people = int(M)

        shirts = []
        for _ in range(num_people):
            index += 1
            person_preferences = get_beecrowd_input(input_test_cases, index, bc_run).split()
            shirts_preferences = [shirt_size_map[person_preferences[0]], shirt_size_map[person_preferences[1]]]
            shirts.append(shirts_preferences)
        structured_test_case = {
            'num_shirts': num_shirts,
            'num_people': num_people,
            'shirts': shirts
        }
        structured_test_cases.append(structured_test_case)

    return structured_test_cases

def minha_camiseta_me_serve(test_cases):
    def edmonds_karp_algorithm(adjency_matrix_graph):
        def bfs(graph, source_index, terminal_index, parents_list):
            visited = [False] * len(graph)
            queue = [source_index]
            visited[source_index] = True

            while queue:
                visiting = queue.pop(0)
                for index, value in enumerate(graph[visiting]):
                    if value > 0 and not visited[index]:
                        queue.append(index)
                        visited[index] = True
                        # print(visited)
                        parents_list[index] = visiting
            # print('finalBFS:', visited, terminal_index, visited[terminal_index])
            return visited[terminal_index]

        def edmonds_karp(graph, source_index, terminal_index):
            # print('e', end='')
            parent = [-1] * len(graph)

            while bfs(graph, source_index, terminal_index, parent):
                path_flow = float('inf')
                flow_terminal = terminal_index
                while flow_terminal != source_index:  # Vai de trás pra frente pegando a menor capacidade
                    # print('F', end='')
                    path_flow = min(path_flow, graph[parent[flow_terminal]][flow_terminal])
                    flow_terminal = parent[flow_terminal]
                
                update_terminal = terminal_index
                while update_terminal != source_index:
                    u = parent[update_terminal]
                    # print('U', end='')
                    graph[u][update_terminal] -= path_flow
                    graph[update_terminal][u] += path_flow
                    update_terminal = parent[update_terminal]
                
                
                msg = '\n'
                for line in adjency_matrix_graph:
                    msg += str(line) + '\n'
                # print(msg)
            return graph
        return edmonds_karp(adjency_matrix_graph, 0, len(adjency_matrix_graph) - 1)
    
    def compute_test_case(test_case):
        """
        - Convert voluntários into nodes with arcs to the sizes they like
            - Each arch has capacity 1
        - Add source node with arcs to voluntários
            - Each arch has capacity 1
        - Add sink node with arcs from all sizes
            - Each arch has capacity test_case['num_shirts']/6
        - Run Edmonds-Karp algorithm
        """
        num_shirts = test_case['num_shirts']
        num_people = test_case['num_people']
        shirts = test_case['shirts']
        vertex_number = 1 + num_people + 6 + 1 # source + voluntários + shirts + sink
        
        graph_capacities = {
            'source': [[1 if 0 < index and index < num_people + 1 else 0 for index in range(vertex_number)]],
            'people_preferences': [],
            'shirts_limit': [],
            'sink': [[0 for _ in range(vertex_number)]],
        }
        
        for shirt_preference in shirts: # Filling people preferences
            adjency_matrix_line = [0] * vertex_number
            for shirt in shirt_preference:
                adjency_matrix_line[1 + num_people + shirt] = 1
            graph_capacities['people_preferences'].append(adjency_matrix_line)
            
        for _ in range(6): # Filling shirts limit
            adjency_matrix_line = [0] * vertex_number
            adjency_matrix_line[-1] = num_shirts // 6
            graph_capacities['shirts_limit'].append(adjency_matrix_line)

        # print(test_case)
        # print('num_shirts', num_shirts//6)
        # for key in graph_capacities:
        #     print(key, len(graph_capacities[key]))
        #     for row in graph_capacities[key]:
        #         print(row)
    
        adjency_matrix = graph_capacities['source'] + graph_capacities['people_preferences'] + graph_capacities['shirts_limit'] + graph_capacities['sink']
        # flux_adjency_matrix = adjency_matrix
        flux_adjency_matrix = edmonds_karp_algorithm(adjency_matrix)
        # for line in flux_adjency_matrix:
            # print(line)
        output_flux = sum(flux_adjency_matrix[-1])
        return 'YES' if output_flux == num_people else 'NO'
    
    return [compute_test_case(test_case) for test_case in test_cases]

def format_output(output_dados):
    return '\n'.join(output_dados)

def main():
    """ Main function """
    input_dados = get_input(bc_run=False)
    output_dados = minha_camiseta_me_serve(input_dados)
    formatted_output = format_output(output_dados)
    print(formatted_output)

main()
```

```python
""" Decoding Scraped Data """

def testing():
    def get_merged_data(beecrowd_input_scraped):
        merged_data = ''
        for key in beecrowd_input_scraped:
            merged_data += beecrowd_input_scraped[key][0]
            merged_data += beecrowd_input_scraped[key][1]
        return merged_data
    def decode_scraped_data(merged_data):
        def RLE_decode(input_string):
            decoded_string = ''
            count = ''
            for char in input_string:
                if char.isdigit():
                    count += char
                else:
                    if count == '':
                        count = '1'
                    decoded_string += int(count) * char
                    count = ''
            return decoded_string
        
        def de_charify_number(char):
            return int(str(ord(char) - ord('@')))

        def de_transform_requests(requests_string):
            def split_integer(integer):
                # Receives an integer and converts it into a list of multiples of 2 that when summed result in the integer
                binary = bin(integer)[2:]
                binary = binary[::-1]
                binary_list = []
                for i, bit in enumerate(binary):
                    if bit == '1':
                        binary_list.append(2**i)
                return binary_list

            def translate_binary_into_shirt(binary_list):
                shirt_size_map = { 1: 'XS', 2: 'S', 4: 'M', 8: 'L', 16: 'XL', 32: 'XXL' }
                shirt_list = []
                for binary in binary_list:
                    shirt_list.append(shirt_size_map[binary])
                return shirt_list
            
            all_requests = requests_string.split('|')
            
            separated_requests = []
            for request_group in all_requests:
                individual_requests = []
                for letter in request_group:
                    value = de_charify_number(letter)
                    splitted_integer_list = split_integer(value)
                    translated_shirt_list = translate_binary_into_shirt(splitted_integer_list)
                    individual_requests.append(translated_shirt_list)
                separated_requests.append(individual_requests)
            return separated_requests
        
        def de_transform_shirts(shirts_string):
            shirts_values = []
            for letter in shirts_string:
                value = de_charify_number(letter)
                complete_value = 6*value
                shirts_values.append(complete_value)
            return shirts_values
        
        def join_into_test_cases(shirts, requests):
            test_cases = []
            for i, request in enumerate(requests):
                test_case = {
                    'num_shirts': str(shirts[i]),
                    'num_people': str(len(request)),
                    'requests': request
                }
                test_cases.append(test_case)
            return test_cases
        
        RLE_decoded = RLE_decode(merged_data)
        shirts, requests_string = RLE_decoded.split(':')
        
        clean_shirts = de_transform_shirts(shirts)
        clean_requests = de_transform_requests(requests_string)
        test_cases = join_into_test_cases(clean_shirts, clean_requests)
        
        return test_cases

    def format_output(decoded):
        formatted_output = str(len(decoded)) + '\n'
        for test_case in decoded:
            formatted_output += test_case['num_shirts'] + ' ' + test_case['num_people'] + '\n'
            for request in test_case['requests']:
                formatted_output += request[0] + ' ' + request[1] + '\n'
        return formatted_output[:-2]

    merged_data = get_merged_data(beecrowd_input_scraped)
    # print(merged_data)
    test_cases = decode_scraped_data(merged_data)
    output_format = format_output(test_cases)
    print(output_format)
    

testing()
```
