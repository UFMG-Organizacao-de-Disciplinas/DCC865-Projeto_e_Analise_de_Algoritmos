## [beecrowd | 1550 - Inversão](https://judge.beecrowd.com/pt/problems/view/1550)

Por Gabriel Dalalio, ITA BR Brazil

Timelimit: 3

Pedro é um garoto curioso que gostava de eletrônica. Certo dia, o menino estava mexendo no laboratório de sua escola e encontrou uma caixa cheia de pequenos aparelhos eletrônicos feitos por outros alunos em anos anteriores.

Dentro dessa caixa havia um aparelho que possuía apenas um visor e dois botões. Esse visor apresentava um número inteiro. Mexendo nos botões, Pedro descobriu para que servia cada um deles. O primeiro botão adicionava uma unidade ao número no visor. O segundo botão invertia os dígitos do número, por exemplo, 123 invertido resulta em 321 e 150 invertido resulta em 51 (ignora-se os zeros a esquerda).

Inicialmente, o visor apresentava o número A. Após a descoberta da função dos botões, Pedro quer saber como fazer o número do visor mudar de A para um número maior igual a B. O seu trabalho nesse problema é ajudar Pedro a descobrir qual é o número mínimo de apertos de botão para que o número no visor passe a ser igual a B.

### Entrada - beecrowd | 1550

A entrada é iniciada por um inteiro $T, 0 < T ≤ 500$, que indica a quantidade de casos de teste a ser processados. Segue-se $T$ linhas cada uma contendo dois inteiros $A$ e $B$, $0 < A < B < 10000$, indicando respectivamente o número inicial no visor e o número que deve ser mostrado no visor depois de apertar os botões.

### Saída - beecrowd | 1550

Para cada caso de teste, o programa deve imprimir um inteiro indicando o número mínimo de apertos de botão para que o número do visor passe de $A$ para $B$.

| 1 9                | 8                |
| 100 301            | 4                |
| 808 909            | 3                |

```python
""" Invertendo e revertendo """
def get_input():
    def convert_input_to_list_of_integers(input_data_item):
        return list(map(int, input_data_item.split()))
    
    def get_beecrowd_input(input_data, index, is_running_on_beecrowd=False):
        """This function gets the input data from the beecrowd website"""
        return input() if is_running_on_beecrowd else input_data[index]
    
    input_data = []
    input_data = uDebug['diogeres']

    index = 0
    [quant_tests] = convert_input_to_list_of_integers(get_beecrowd_input(input_data, index))
    in_out = []
    for _ in range(quant_tests):
        index += 1
        test_case = convert_input_to_list_of_integers(get_beecrowd_input(input_data, index))
        in_out.append(test_case)

    data_input = in_out

    return data_input


def inversao(input_dados):
    def inverte(valor_a_ser_invertido, is_output=False):
        last_digit = valor_a_ser_invertido % 10
        if is_output and last_digit == 0:
            return valor_a_ser_invertido
        
        inverted_value = 0
        while valor_a_ser_invertido > 0:
            inverted_value = inverted_value * 10 + valor_a_ser_invertido % 10
            valor_a_ser_invertido = valor_a_ser_invertido // 10
        return inverted_value

    def soma(valor_a_ser_somado):
        return valor_a_ser_somado + 1

    def subtrai(valor_a_ser_subtraido):
        return valor_a_ser_subtraido - 1 if valor_a_ser_subtraido > 0 else 0

    def get_value_depth(value, depth):
        return {'value': value, 'depth': depth}

    def append_new_visitor(visited_values, nodes_to_visit, input_values_to_visit, visiting_node, new_node_value):
        if new_node_value not in visited_values and new_node_value not in input_values_to_visit:
            nodes_to_visit.append(get_value_depth(new_node_value, visiting_node["depth"] + 1))
            input_values_to_visit.add(new_node_value)

    def finish_visit(visited_values, nodes_to_visit):
        for visiting_node in nodes_to_visit:
            visited_values[visiting_node["value"]] = visiting_node["depth"]

    def get_minimum_depth(test_case):
        input_value = test_case[0]
        input_values_to_visit = {input_value}
        input_nodes_to_visit = [get_value_depth(input_value, 0)]
        visited_inputs = dict()
        
        output_value = test_case[1]
        output_values_to_visit = {output_value}
        output_nodes_to_visit = [get_value_depth(output_value, 0)]
        visited_outputs = dict()

        intersect = {}
        
        while not intersect:
            # INPUT PART
            visiting_input = input_nodes_to_visit.pop(0)
            input_values_to_visit.remove(visiting_input["value"])
            visited_inputs[visiting_input["value"]] = visiting_input["depth"]

            append_new_visitor(visited_inputs, input_nodes_to_visit, input_values_to_visit, visiting_input, soma(visiting_input['value']))
            append_new_visitor(visited_inputs, input_nodes_to_visit, input_values_to_visit, visiting_input, inverte(visiting_input['value']))

            # OUTPUT PART
            visiting_output = output_nodes_to_visit.pop(0)
            output_values_to_visit.remove(visiting_output["value"])
            visited_outputs[visiting_output["value"]] = visiting_output["depth"]

            append_new_visitor(visited_outputs, output_nodes_to_visit, output_values_to_visit, visiting_output, subtrai(visiting_output['value']))
            append_new_visitor(visited_outputs, output_nodes_to_visit, output_values_to_visit, visiting_output, inverte(visiting_output['value'], True))

            intersect = set(visited_inputs.keys()).intersection(visited_outputs)

        finish_visit(visited_inputs, input_nodes_to_visit)
        finish_visit(visited_outputs, output_nodes_to_visit)
        
        final_intersect = set(visited_inputs.keys()).intersection(visited_outputs)

        minimum_depth = min([visited_inputs[value] + visited_outputs[value] for value in final_intersect])
        
        return minimum_depth

    results = []
    for test_case in input_dados:
        results.append(get_minimum_depth(test_case))
    return results


def format_output(output_dados):
    """ Format output """
    return "\n".join(map(str, output_dados))


def main():
    """ Main function """
    input_dados = get_input()
    output_dados = inversao(input_dados)
    formatted_output = format_output(output_dados)
    print(formatted_output)


main()
```
